```python
import random
import matplotlib.pyplot as plt
```

```python
class AdSlot:
    def __init__(self, name, true_ctr):
        self.name = name
        self.true_ctr = true_ctr
        self.impressions = 0
        self.clicks = 0
        self.click_history = []
        self.ctr_history = []

    def show_ad(self):
        self.impressions += 1
        if random.random() < self.true_ctr:
            self.clicks += 1
            self.click_history.append(self.clicks)
            self.ctr_history.append(self.get_ctr())
            return 1
        self.click_history.append(self.clicks)
        self.ctr_history.append(self.get_ctr())
        return 0

    def get_ctr(self):
        if self.impressions == 0:
            return 0
        return self.clicks / self.impressions

```

```python
class EpsilonGreedy:
    def __init__(self, ad_slots, epsilon=0.1):
        self.ad_slots = ad_slots
        self.epsilon = epsilon

    def select_ad_slot(self):
        if random.random() < self.epsilon:
            return random.choice(self.ad_slots)
        else:
            return max(self.ad_slots, key=lambda x: x.get_ctr())

    def run_simulation(self, num_impressions):
        total_clicks = 0
        for _ in range(num_impressions):
            selected_slot = self.select_ad_slot()
            reward = selected_slot.show_ad()
            total_clicks += reward

        print(f"Total clicks: {total_clicks}")
        print("Final CTRs:")
        for slot in self.ad_slots:
            print(f"{slot.name}: {slot.get_ctr():.4f}")

        self.plot_results(num_impressions)

    def plot_results(self, num_impressions):
        plt.figure(figsize=(12, 5))
```

```python
class EpsilonGreedy:
    def __init__(self, ad_slots, epsilon=0.1):
        self.ad_slots = ad_slots
        self.epsilon = epsilon

    def select_ad_slot(self):
        if random.random() < self.epsilon:
            return random.choice(self.ad_slots)
        else:
            return max(self.ad_slots, key=lambda x: x.get_ctr())

    def run_simulation(self, num_impressions):
        total_clicks = 0
        for _ in range(num_impressions):
            selected_slot = self.select_ad_slot()
            reward = selected_slot.show_ad()
            total_clicks += reward

        print(f"Total clicks: {total_clicks}")
        print("Final CTRs:")
        for slot in self.ad_slots:
            print(f"{slot.name}: {slot.get_ctr():.4f}")

        self.plot_results(num_impressions)

    def plot_results(self, num_impressions):
        plt.figure(figsize=(12, 5))

        # Plot cumulative clicks
        plt.subplot(1, 2, 1)
        for slot in self.ad_slots:
            plt.plot(range(len(slot.click_history)), slot.click_history, label=slot.name) # Changed from range(num_impressions) to range(len(slot.click_history)) so that the x-axis has the same number of values as the y-axis
        plt.title('Cumulative Clicks Over Time')
        plt.xlabel('Impressions')
        plt.ylabel('Cumulative Clicks')
        plt.legend()

        # Plot CTR over time
        plt.subplot(1, 2, 2)
        for slot in self.ad_slots:
            plt.plot(range(len(slot.ctr_history)), slot.ctr_history, label=slot.name) # Changed from range(num_impressions) to range(len(slot.ctr_history)) so that the x-axis has the same number of values as the y-axis
        plt.title('Estimated CTR Over Time')
        plt.xlabel('Impressions')
        plt.ylabel('Estimated CTR')
        plt.legend()

        plt.tight_layout()
        plt.show()
```

```python
ad_slots = [
    AdSlot("Top Banner", 0.05),
    AdSlot("Sidebar", 0.02),
    AdSlot("Footer", 0.01),
    AdSlot("Pop-up", 0.03)
]

# Run the simulation
epsilon_greedy = EpsilonGreedy(ad_slots)
epsilon_greedy.run_simulation(1000)
```
