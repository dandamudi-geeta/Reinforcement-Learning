```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class NonStationaryMAB:
    def __init__(self, n_arms, change_interval):
        self.n_arms = n_arms
        self.change_interval = change_interval
        self.means = np.random.rand(n_arms)  # Initial means for each arm
        self.total_rewards = np.zeros(n_arms)  # Total rewards for each arm
        self.counts = np.zeros(n_arms)  # Counts for each arm pulled
        self.regret_history = []  # Track regrets over time
        self.best_arm = np.argmax(self.means)  # Track the best arm
```

```python
def pull(self, arm):
        # Simulate a reward based on the arm's current mean
        reward = np.random.binomial(1, self.means[arm])
        return reward
```

```python
def update_means(self):
        # Change the means every 'change_interval'
        if len(self.regret_history) % self.change_interval == 0 and len(self.regret_history) > 0:
            self.means = np.random.rand(self.n_arms)  # New means for the arms
            self.best_arm = np.argmax(self.means)  # Update the best arm
```

```python


    def run(self, n_rounds, epsilon):
        rewards = np.zeros((n_rounds, self.n_arms))  # Array to track rewards over rounds

        for round in range(n_rounds):
            self.update_means()  # Update means according to the non-stationary setting

            # Epsilon-greedy strategy
            if np.random.rand() < epsilon:
                arm = np.random.randint(self.n_arms)  # Explore
            else:
                # Exploit: choose the arm with the highest average reward
                arm = np.argmax(self.total_rewards / (self.counts + 1e-5))  # Avoid division by zero

            reward = self.pull(arm)  # Get the reward from the selected arm

            # Update total rewards and counts
            self.total_rewards[arm] += reward
            self.counts[arm] += 1
            rewards[round, :] = self.total_rewards / (self.counts + 1e-5)  # Record the average rewards

            # Record the regret (how much reward we lost by not choosing the best arm)
            regret = self.total_rewards.sum() - self.total_rewards[self.best_arm]
            self.regret_history.append(regret)

        return rewards
```
